### Exercise 5

**Question.** What are the advantages and disadvantages of a very flexible (versus a less flexible) approach for regression or classification? Under what circumstances might a more flexible approach be preferred to a less flexible approach? When might a less flexible approach be preferred?

**Answer.** A very flexible model can capture big non-linear problems. This means that we can approximate very complex problems in a good way. On the other hand, it is more challenging to interpret and has high variance. If we have a huge training set, it is better to use a highly flexible model. If not, better use a less flexible approach. A less flexible approach is preferred when we have small data and know that the problem is linear (or quite linear).

### Exercise 6

***Question.*** Describe the differences between a parametric and a non-parametric statistical learning approach. What are the advantages of a para- metric approach to regression or classification (as opposed to a non- parametric approach)? What are its disadvantages?

***Answer.*** 
- A parametric approach reduces the problem of estimating f down to one of estimating a set of parameters because it assumes a form for f.
- A non-parametric approach does not assume a patricular form of f and so requires a very large sample to accurately estimate f.
- The advantages of a parametric approach to regression or classification are the simplifying of modeling f to a few parameters and not as many observations are required compared to a non-parametric approach.
- The disadvantages of a parametric approach to regression or classification are a potentially inaccurate estimate f if the form of f assumed is wrong or to overfit the observations if more flexible models are used.

**Question.** The table below provides a training data set containing six observa- tions, three predictors, and one qualitative response variable.

```python
table = pd.DataFrame(data = {'X_1': [0,2,0,0,-1,1], 
                             'X_2': [3,0,1,1,0,1], 
                             'X_3': [0,0,3,2,1,1], 
                             'Y': ['Red','Red','Red','Green', 'Green', 'Red']}, 
                    index = [1,2,3,4,5,6])
display(table)
```

Suppose we wish to use this data set to make a prediction for Y when $X_1 = X_2 = X_3 = 0$ using K-nearest neighbors.
- (a) Compute the Euclidean distance between each observation and the test point, $X_1 = X_2 = X_3 = 0$
- (b) What is our prediction with K = 1? Why?
- (c) What is our prediction with K = 3? Why?
- (d) If the Bayes decision boundary in this problem is highly non- linear, then would we expect the best value for K to be large or small? Why?

**Answer.** 
- (a) We have that:

```python
for j in table.index:
    euclidian_distance = np.sum(np.array(table.loc[j])[:-1]**2)**0.5
    euclidian_distance = round(euclidian_distance, 2)
    print(f"The euclidian distance from observation {j} to the test point (0, 0, 0) is {euclidian_distance}.")
    
```

- (b) The prediction will be Green ,because observation 5 is the closest to the test point.
- (c) The prediction will be Red, becase we have three observation in the neighborhood of the test point and two of them are Red
